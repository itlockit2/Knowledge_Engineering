```
# 사진을 화면에 출력
import os
from IPython.display import display
from PIL import Image
IMAGE_PATH = os.path.join(os.getcwd(),'drive','My Drive','Knowledge_engineering', 'chap_5')

def imageOpen(fileName):
  im = Image.open(os.path.join(IMAGE_PATH,fileName))
  resize_width = 600
  width, height = im.size
  resize_height = resize_width * height / width
  im.thumbnail((resize_width,resize_height))
  return im
```

```
display(imageOpen('5-3.png'))
```
